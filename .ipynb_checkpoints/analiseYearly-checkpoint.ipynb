```python
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
#plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})
```

```python
df = pd.read_csv('2000_2021_Brazilian_North_Region_hourly_MWmed_hydroelectric_power_series_train.csv', parse_dates=['data'], index_col='data')
```

```python
def yearly_lineplot(series, years):
    series['hora'] = [d.strftime('%H') for d in series.index]
    series['dia'] = [d.strftime('%d') for d in series.index]
    series['mes'] = [d.strftime('%m') for d in series.index]
    series['ano'] =[d.strftime('%Y') for d in series.index]
    series['mes dia hora'] = series['mes'] + "-" + series['dia'] + "-" + series['hora']

    anos = df['ano'].unique()
    
    count = 0
    for i, a in enumerate(anos):

        plt.plot('mes dia hora','energia', data=series.loc[series.ano==a,:], label=a)
        plt.legend(bbox_to_anchor=(1,1), loc="upper left")
        count = count + 1
        if count == years:
            break

    plt.show()
```

```python
yearly_lineplot(df,2)
```
