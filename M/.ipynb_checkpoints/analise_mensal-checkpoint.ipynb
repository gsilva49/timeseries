```python
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def monthly_lineplot(series, months):
    series['dia'] = [d.strftime('%d') for d in series.index]
    series['mes_ano'] = [d.strftime('%b-%Y') for d in series.index]

    meses = series['mes_ano'].unique()
    dia = series['dia'].unique()

    count = 0
    for i, m in enumerate(meses):

        plt.plot('dia','energia', data=series.loc[series.mes_ano==m,:], label=m)
        plt.legend(bbox_to_anchor=(1,1), loc="upper left")
        count = count + 1
        if count == months:
            break
    
    plt.xticks(dia[::5], dia[::5], rotation='vertical')
    
    plt.show()
```
