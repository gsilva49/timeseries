```python
# detrend a time series using differencing
from pandas import read_csv
from datetime import datetime
from pandas import Series
from matplotlib import pyplot
```

```python
# invert a differenced series 
def invert_difference(train, diff):
    series = []
    # add first value
    series.append(train[0])
    for idx in range(len(diff)):
        # sum the current value and its prior value 
        series.append(diff[idx] + train[idx])
    series = Series(series)
    series.index = train.index
    return series
```

```python
# difference a series
def difference(train):
    diff = list()
    # lose first value
    for i in range(1, len(train)):
        # substract the current value and its prior value 
        value = train[i] - train[i - 1]
        diff.append(value)
    diff = Series(diff)
    diff.index = train.index[1:]
    return diff
```

```python
# format datetime
def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
```

```python
# load and plot a series
train = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True, squeeze=True, date_parser=parser)
pyplot.plot(train)
pyplot.show()
```

```python
# difference and plot a series
diff = difference(train)
pyplot.plot(diff)
pyplot.show()
```

```python
# invert and plot differenced series
inv_diff = invert_difference(train, diff)
pyplot.plot(inv_diff)
pyplot.show()
```
